```python
# import the libraries required to perform the regression

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
# load the dataset

df = pd.read_csv(r"D:\world happiness report\2015.csv")
df.columns
```

```python
# arranging the variables and target

X = df.drop(df.iloc[:,:4],axis=1)
y = df.iloc[:,2:4]
```

```python
# spliting the data into train and test 

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
# fitting the data to the model -> training the model

regressor = DecisionTreeRegressor(criterion='squared_error',max_depth=3,random_state=1)
regressor.fit(X_train,y_train)
```

```python
#predicting the values with the model which is trained on training dataset

y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error

# testing the model score 
print("Test score %3f" %regressor.score(X_test, y_test))
print("Train score %3f" %regressor.score(X_train, y_train))
```

```python
# metrics used to evaluate the performance of the model

mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
print("mse %3f" %mse)
print("mae %3f" %mae)
```
